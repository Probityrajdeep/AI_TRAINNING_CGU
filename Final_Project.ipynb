```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv("/content/placementdata.csv")
df.head()
```

```python
df['CGPA'].describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop("StudentID", axis=1, inplace=True)
```

```python
df["ExtracurricularActivities"] = df["ExtracurricularActivities"].map({"Yes": 1, "No": 0})
df["PlacementTraining"] = df["PlacementTraining"].map({"Yes": 1, "No": 0})

df["PlacementStatus"] = df["PlacementStatus"].map({
    "Placed": 1,
    "NotPlaced": 0
})
```

```python
print(df["PlacementStatus"].unique())  # must be [0 1]
```

```python
X = df.drop("PlacementStatus", axis=1)
y = df["PlacementStatus"]
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.25, random_state=0, stratify=y)
```

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
new_student_original_cgpa = 5.5   # change this value as needed
```

```python
new_student = np.array([[
    new_student_original_cgpa,  # CGPA
    78,    # SSC_Marks
    82,    # HSC_Marks
    1,     # Internships
    3,     # Projects
    2,     # Workshops/Certifications
    85,    # AptitudeTestScore
    4,     # SoftSkillsRating
    1,     # ExtracurricularActivities
    1      # PlacementTraining
]])
```

```python
new_student_scaled = scaler.transform(new_student)

if new_student_original_cgpa < 6.0:
    print("Student is likely to be NOT PLACED (CGPA below cutoff)")
else:
    prediction = model.predict(new_student_scaled)
    if prediction[0] == 1:
        print("Student is likely to be PLACED")
    else:
        print("Student is likely to be NOT PLACED")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight="balanced"
)

rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": rf_model.feature_importances_
}).sort_values(by="Importance", ascending=False)

print(importance)
```

```python
import streamlit as st
import numpy as np
import joblib
import pandas as pd

# Load the saved model and scaler
rf_model = joblib.load('random_forest_model.pkl')
scaler = joblib.load('scaler.pkl')

def predict_placement(new_student_data):
    # Extract CGPA for cutoff rule
    new_student_original_cgpa = new_student_data[0]

    if new_student_original_cgpa < 6.0:
        return "NOT PLACED (CGPA below cutoff)"
    else:
        # Scale the new student data
        new_student_scaled = scaler.transform([new_student_data])
        prediction = rf_model.predict(new_student_scaled)
        if prediction[0] == 1:
            return "PLACED"
        else:
            return "NOT PLACED"

# Streamlit App
st.set_page_config(page_title="Student Placement Prediction", layout="centered")
st.title("🎓 Student Placement Prediction")
st.markdown("---")

st.header("Enter Student Details")

# Input fields for student features
with st.form("placement_form"):
    col1, col2 = st.columns(2)
    with col1:
        cgpa = st.number_input("CGPA (6.5 - 9.1)", min_value=6.5, max_value=9.1, value=7.5, step=0.1)
        ssc_marks = st.number_input("SSC Marks (0-100)", min_value=0, max_value=100, value=70, step=1)
        hsc_marks = st.number_input("HSC Marks (0-100)", min_value=0, max_value=100, value=75, step=1)
        internships = st.slider("Number of Internships", min_value=0, max_value=5, value=1)
        projects = st.slider("Number of Projects", min_value=0, max_value=5, value=2)
    with col2:
        workshops = st.slider("Workshops/Certifications", min_value=0, max_value=5, value=1)
        aptitude_score = st.number_input("Aptitude Test Score (0-100)", min_value=0, max_value=100, value=70, step=1)
        soft_skills = st.number_input("Soft Skills Rating (0-5)", min_value=0.0, max_value=5.0, value=3.5, step=0.1)
        extracurricular = st.radio("Extracurricular Activities", options=["Yes", "No"])
        placement_training = st.radio("Placement Training", options=["Yes", "No"])

    extracurricular_val = 1 if extracurricular == "Yes" else 0
    placement_training_val = 1 if placement_training == "Yes" else 0

    submitted = st.form_submit_button("Predict Placement")

    if submitted:
        new_student_input = [
            cgpa,
            internships,
            projects,
            workshops,
            aptitude_score,
            soft_skills,
            extracurricular_val,
            placement_training_val,
            ssc_marks,
            hsc_marks
        ]

        # Features for the model should be in the order: CGPA, Internships, Projects, Workshops/Certifications, AptitudeTestScore, SoftSkillsRating, ExtracurricularActivities, PlacementTraining, SSC_Marks, HSC_Marks
        # The order in new_student_input above needs to match this.
        # Let's re-confirm the order from X.columns
        # X.columns: Index(['CGPA', 'Internships', 'Projects', 'Workshops/Certifications', 'AptitudeTestScore', 'SoftSkillsRating', 'ExtracurricularActivities', 'PlacementTraining', 'SSC_Marks', 'HSC_Marks'], dtype='object')

        # Re-ordering to match the model's training features
        ordered_input = [
            cgpa, # CGPA
            internships, # Internships
            projects, # Projects
            workshops, # Workshops/Certifications
            aptitude_score, # AptitudeTestScore
            soft_skills, # SoftSkillsRating
            extracurricular_val, # ExtracurricularActivities
            placement_training_val, # PlacementTraining
            ssc_marks, # SSC_Marks
            hsc_marks  # HSC_Marks
        ]

        prediction_result = predict_placement(ordered_input)

        st.subheader("Prediction Result:")
        if "PLACED" in prediction_result:
            st.success(f"🎉 Student is likely to be {prediction_result}")
        else:
            st.error(f"❌ Student is likely to be {prediction_result}")

st.markdown("--- ")

st.subheader("Feature Importance (from Random Forest Model)")

# Assuming `importance` DataFrame is available from the notebook
# If not, you'd calculate it within the app or load it if saved.
# For now, let's just display a mock-up or the last computed one.

# You can uncomment and run this part if you save the feature importance as a CSV or similar
# importance_df = pd.read_csv('feature_importance.csv') # Example if saved
# st.dataframe(importance_df)

# Using the importance data available from the notebook's kernel state
# Feature names in the same order as X.columns
feature_names = ['CGPA', 'Internships', 'Projects', 'Workshops/Certifications', 'AptitudeTestScore', 'SoftSkillsRating', 'ExtracurricularActivities', 'PlacementTraining', 'SSC_Marks', 'HSC_Marks']
feature_importances = rf_model.feature_importances_

importance_df = pd.DataFrame({
    "Feature": feature_names,
    "Importance": feature_importances
}).sort_values(by="Importance", ascending=False)

st.dataframe(importance_df)
```
